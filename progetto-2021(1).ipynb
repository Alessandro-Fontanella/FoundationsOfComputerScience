### 2. For each book in the *Books* dataset, compute its average rating.

```python
## Provo a uniformare gli ISBN, poichè molte entry sono errate
df_ratings["ISBN"] =df_ratings["ISBN"].apply(lambda x: x.strip().strip("\'").strip("\\").strip('\"').strip("\#").strip("(").strip(")").strip('*').strip('+').strip('/').strip(',.').strip('.').strip('/'))
```

```python
df_ratings['Book-Rating'] = df_ratings['Book-Rating'].astype('float64')
```

```python
book_groupby = df_ratings.groupby("ISBN")
```

```python
### Calcolo la media sul raggruppamento per ISBN dei libri
average_book_rating = book_groupby["Book-Rating"].mean()
```

```python
average_book_rating = average_book_rating.to_frame()
```

```python
average_book_rating
```

```python
### Left join tra books e le medie calcolate
df_books = df_books.join(average_book_rating, on="ISBN")
```

### 3. For each book in the *GoodBooks* dataset, compute its average rating.

```python
pd.set_option('display.max_columns', None)
df_goodbooks
```

```python
### Imposto il tipo corretto per poter eseguire la somma
df_goodbooks['ratings_1'] = df_goodbooks['ratings_1'].astype(int)
df_goodbooks['ratings_2'] = df_goodbooks['ratings_2'].astype(int)
df_goodbooks['ratings_3'] = df_goodbooks['ratings_3'].astype(int)
df_goodbooks['ratings_4'] = df_goodbooks['ratings_4'].astype(int)
df_goodbooks['ratings_5'] = df_goodbooks['ratings_5'].astype(int)
df_goodbooks['work_ratings_count'] = df_goodbooks['work_ratings_count'].astype(int)
```

### 7. For each book in `goodbooks`, compute the list of its authors. Assuming that the number of reviews with a text (column `work_text_reviews_count`) is split equally among all authors, find for each authors the total number of reviews with a text. We will call this quantity the *shared number of reviews with a text*.

```python
### Costruisco una lista di autori splittando la colonna author_list
df_goodbooks['author_list'] = df_goodbooks.authors.apply(lambda x: x.split(','))
```

```python
df_goodbooks.head()
```

```python
### Calcolo il numero di autori per libro
df_goodbooks['author_number'] = df_goodbooks.author_list.apply(lambda  x:len(x))
```

```python
df_goodbooks['work_text_reviews_count'] = df_goodbooks['work_text_reviews_count'].astype(int)
```

```python
### Calcolo la quota di recensioni condivise
df_goodbooks['shared_number'] = df_goodbooks['work_text_reviews_count'] / df_goodbooks['author_number']
```

```python
df_goodbooks.head()
```

```python
### Esplodo il dataset per ottenere per ogni libro un singolo autore e non più la lista di autori
l = df_goodbooks.apply(pd.Series.explode).reset_index()
l.head()
```

```python
group_autori= l.groupby(['author_list'], as_index=False)
group_autori = group_autori["shared_number"].sum()
group_autori.head()
```

### 8. For each year of publication, determine the author that has the largest value of the shared number of reviews with a text.

```python
df_goodbooks.head()
```

```python
l = df_goodbooks.apply(pd.Series.explode).reset_index()
l.head()
```

```python
### Recupero la posizione per ogni anno di pubblicazione dell'autore con shared number massimo
result = group_autori.loc[group_autori.groupby(['original_publication_year'])['shared_number'].idxmax()]
result
```
